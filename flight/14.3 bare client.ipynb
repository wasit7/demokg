```python
import rstr
import pandas as pd
n=int(10e6)
d={
    'a':[rstr.xeger(r'(\w{10,20})') for i in range(n)],
    'b':[rstr.xeger(r'([A-Z]\d{20})') for i in range(n)],
    'c':[rstr.xeger(r'([A-Z]\d{10})') for i in range(n)],
}
df=pd.DataFrame(d)
table=pa.table(d)
```

```python
upload_descriptor = pa.flight.FlightDescriptor.for_path('test.parquet')
writer, flight = client.do_put(upload_descriptor, table.schema)
writer.write_table(table)
writer.close()
```
