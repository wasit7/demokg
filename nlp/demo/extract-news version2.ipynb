```python
import threading
```

```python
def new2impact(_df):
    global output
    _data = _df[['article_title', 'article_content', 'article_image_url', 'url']]\
            .rename(columns={'article_title':'name', 'article_content':'detail'})\
            .to_dict('records')
    for row in tqdm(_data):
        try:
            r=openai.ChatCompletion.create(
              model="gpt-3.5-turbo",
              messages=[
                    {"role": "system", "content": instruction},
                    {"role": "user", "content": input_article_from_user},
                    {"role": "assistant", "content": output_analysis_str},
                    {"role": "user", "content": row['detail']}
              ]
            )
            row.update({
                'impact': _parser(r)
            })
        except Exception as e:
            print(e)
            continue
    output += _data

def compute_partitions(df, n_thread):
    n_size = len(df)//n_thread + 1
    return [(elm*n_size, (elm+1)*n_size) for elm in range(n_thread)]
```

```python
threads = []
output = []
for i,j in compute_partitions(df, 4):
    _df = df[i:j]
    t = threading.Thread(target=new2impact, args=(_df,))
    t.start()
    threads.append(t)
for thread in threads:
    thread.join()
```
