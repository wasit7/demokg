```python
import spacy_thai
# import graphviz
import deplacy
import pandas as pd
nlp=spacy_thai.load()
```

```python
doc=nlp("กรใช้ภาษาไทยและเฉินใช้ภาษาจีน")
deplacy.render(doc,WordRight=True)
df=pd.DataFrame(doc.to_json()['tokens'])
df['text'] = df.apply(lambda x: doc.text[x['start']:x['end']], axis=1)
df
```

```python
doc=nlp("เรามีแม่แมวสีแดงกำลังให้นมลูกเราชอบมันมาก")
deplacy.render(doc,WordRight=True)
df=pd.DataFrame(doc.to_json()['tokens'])
df['text'] = df.apply(lambda x: doc.text[x['start']:x['end']], axis=1)
df
```
